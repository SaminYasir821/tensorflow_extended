```python
import os
from tfx.components import CsvExampleGen
from tfx.orchestration.experimental.interactive.interactive_context import InteractiveContext
import glob

# Setup interactive TFX context
context = InteractiveContext()

# Define the correct data directory path
base_dir = os.getcwd()
data_dir = os.path.join(base_dir, "data")  # Correct path!

print(f"Looking for CSV files in: {data_dir}")

# Debug: print available CSV files
csv_files = glob.glob(os.path.join(data_dir, "*.csv"))
print("Found CSV files:", csv_files)

# # If no CSV files found, raise an error
# if not csv_files:
#     raise FileNotFoundError(f"No CSV files found in {data_dir}")

# Create CsvExampleGen component with input_base parameter
example_gen = CsvExampleGen(input_base=data_dir)

# Run the component
context.run(example_gen)
```
